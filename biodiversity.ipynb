```python
species.fillna('No Intervention', inplace=True)
species.groupby('conservation_status').size()
```

```python
conservationCategory = species[species.conservation_status != 'No Intervention'].groupby(['conservation_status', 'category'])['scientific_name'].count().unstack()

conservationCategory
```

```python
ax = conservationCategory.plot(kind='bar', figsize=(8,6), stacked=True)
ax.set_xlabel('Conservation Status')
ax.set_ylabel('Number of Species');
```

```python
species['is_protected'] = species.conservation_status != 'No Intervention'

category_counts = species.groupby(['category', 'is_protected']).scientific_name.nunique().reset_index().pivot(columns='is_protected', index='category', values='scientific_name').reset_index()
category_counts.columns = ['category', 'not_protected', 'protected']
category_counts
```

```python
category_counts['percents_protected'] = category_counts.protected/(category_counts.protected+category_counts.not_protected)*100

category_counts
```

```python
from scipy.stats import chi2_contingency

contingency1 = [[30, 146],[75, 413]]
chi2_contingency(contingency1)
print(f'First Contingency: {chi2_contingency(contingency1)}')
contingency2 = [[30, 146], [5, 73]]
chi2_contingency(contingency2)
print(f'Second Contingency: {chi2_contingency(contingency2)}')
```

```python
from itertools import chain
import string

def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text

common_Names = species[species.category == 'Mammal'].common_names.apply(remove_punctuations).str.split().tolist()
common_Names[:6]
```

```python
cleanRows = []

for item in common_Names:
    item = list(dict.fromkeys(item))
    cleanRows.append(item)
cleanRows[:6]
```

```python
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in cleanRows))
res[:6]
```

```python
words_counted = []

for i in res:
    x = res.count(i)
    words_counted.append((i, x))
    
pd.DataFrame(set(words_counted), columns=['Word', 'Count']).sort_values('Count', ascending = False).head(10)
```

```python
species['is_bat'] = species.common_names.str.contains(r"\bBat\b", regex = True)
species.head(10)
```

```python
species[species.is_bat]
```

```python
bat_observations = observations.merge(species[species.is_bat])
bat_observations
```

```python
bat_observations.groupby('park_name').observations.sum().reset_index()
```

```python
observations_by_park = bat_observations.groupby(['park_name', 'is_protected']).observations.sum().reset_index()
observations_by_park
```

```python
plt.figure(figsize=(16,4))
sns.barplot(x=observations_by_park.park_name, y = observations_by_park.observations, hue = observations_by_park.is_protected)
plt.xlabel('National Parks')
plt.ylabel('Number of Observations')
plt.title('Observations of Bats')
plt.show()
```
